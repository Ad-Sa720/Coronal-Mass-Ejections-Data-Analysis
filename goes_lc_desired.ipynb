```python
from sunpy.net import Fido, attrs as a
import matplotlib.pyplot as plt
from sunpy import timeseries as ts
```

```python
tstart = "2012-03-07 1:40"
tend = "2012-03-07 02:40"
result = Fido.search(a.Time(tstart, tend),
                     a.Instrument("XRS"))
```

```python
result
```

```python
file_goes = Fido.fetch(result)
```

```python
goes_15 = ts.TimeSeries(file_goes)
```

```python
goes_15.peek()
```

```python
#saving the plots
fig, ax = plt.subplots()
goes_15.plot()
fig.savefig("goes_lc_desired.png")
```
