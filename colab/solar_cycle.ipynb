```python
#download the data
import wget
```

```python
files=["SN_d_tot_V2.0.txt","SN_m_tot_V2.0.txt","SN_ms_tot_V2.0.txt","SN_y_tot_V2.0.txt"]
for f in files:
    url="https://www.sidc.be/silso/DATA/"+f
    wget.download(url,'data/')
    print(url)
```

### Plotting Sunspot number

```python
import matplotlib.pyplot as plt
import numpy as np
```

#### Daily Sunspot number

```python
d = np.loadtxt('data/SN_d_tot_V2.0.txt',skiprows=0)
```

```python
plt.figure(figsize=(15,5))
plt.plot(d[:,3],d[:,4])
plt.xlabel("Time (year)")
plt.ylabel("Daily Sunspot Number")
```

#### Monthly Sunspot number

```python
d = np.loadtxt('data/SN_m_tot_V2.0.txt',skiprows=0)
```

```python
plt.figure(figsize=(15,5))
plt.plot(d[:,2],d[:,3])
plt.xlabel("Time (year)")
plt.ylabel("Monthly Sunspot Number")
```

#### Yearly averged Sunspot number

```python
d = np.loadtxt('data/SN_y_tot_V2.0.txt',skiprows=0)
```

```python
plt.figure(figsize=(15,5))
plt.plot(d[:,0],d[:,1])
plt.xlabel("Time (year)")
plt.ylabel("Yearly Sunspot Number")
```
