## Period of the Solar cycle

```python
import numpy as np
import matplotlib.pyplot as plt 
from scipy.fft import fft
```

```python
d = np.loadtxt('data/SN_m_tot_V2.0.txt')
```

```python
plt.figure(figsize=(15,5))
plt.plot(d[:,2],d[:,3])
```

```python
amp = fft(d[:,3])

N = np.size(amp)
pw = (np.abs(amp[0:N//2]))**2

dt = 1.0/12.
fnyq = 1.0/(2.0*dt)
fr = np.linspace(0.0,fnyq,N//2)
```

```python
plt.plot(fr,pw)
plt.xlim(0.0,0.2)
plt.ylim(0,1e10)
plt.xlabel("Frequency (1/year)")
plt.ylabel("Power (arb. unit)")

ind = np.where((fr>0.05) & (fr<0.2))
im = np.argmax(pw[ind])
mfr = fr[np.min(ind)+im]
print(mfr)

plt.axvline(mfr,color='r',linewidth=2)
print("period = %10.4f year" % (1/mfr))
```
